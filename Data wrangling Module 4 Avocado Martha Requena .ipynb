4. Write a function that takes a row as input and returns a shortened version of the values in the Variety column. The values should consist of the first two words of the variety name, followed by a space and an ellipsis (…). Hint: You’ll need to split the values into a list of words, get the first two words in the list, add the ellipsis to the list, and then join the words in the list.

```python
def shorten_variety(row):
    # Split the value in 'Variety' into a list of words.
    words = row['Variety'].split()
    
    # Get the first two words from the list.
    first_two_words = words[:2]
    
    # Add an ellipsis to the list.
    first_two_words.append('...')
    
    # Join the words in the list into a single string.
    shortened_variety = ' '.join(first_two_words)
    
    return shortened_variety

# Example usage with a DataFrame row:
# Make sure 'ramen' is your DataFrame containing the 'Variety' column
shortened_text = shorten_variety(ramen.iloc[0])
print(shortened_text)
```

5. Apply the function you just wrote to the DataFrame, and store the results in a column named ShortVariety. Then, display the first five rows to be sure this worked.

```python
# Apply the 'shorten_variety' function to each row of the DataFrame 'ramen'
# and store the results in a new column 'ShortVariety'.
ramen['ShortVariety'] = ramen.apply(shorten_variety, axis=1)

# Display the first five rows of the DataFrame to ensure the operation worked correctly.
print(ramen.head())
```
